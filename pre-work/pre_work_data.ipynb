#### **Precios**

```python
# Prices df
path = "data\original\item_prices.csv"

prices_df = pd.read_csv(path, sep=",", encoding="utf-8")

prices_df.head(3)
```

```python
# Hay nulos? Yes
prices_df.isna().sum()
```

```python
# Hay duplicados?
prices_df.duplicated().value_counts()
```

```python
# Coinciden las filas duplicadas con las que yearweek isna?
# Que significa esto? -> Aportan informacin? Creo que no -> Si no tengo la semana no puedo relacionar ese precio con nada
prices_df[prices_df.duplicated()].equals(prices_df[prices_df.duplicated() & prices_df['yearweek'].isna()])
```

```python
# Drop nas
prices_df = prices_df[prices_df['yearweek'].notna()]
```

```python
# prices_df.info()
```

```python
# Convertir la columna 'yearweek' a tipo string y luego separarla en ao y semana
prices_df.loc[:,'yearweek'] = prices_df.loc[:,'yearweek'].astype(str)
```

```python
prices_df['year'] = prices_df.loc[:,'yearweek'].str[:4] # Ao
prices_df['week'] = prices_df.loc[:,'yearweek'].str[4:6] # Semana

# Crear una nueva columna de fecha en el formato year-week
prices_df['year-week'] = prices_df.loc[:,'year'] + '-W' + prices_df.loc[:,'week']

# Convertir la nueva columna de string a un objeto datetime
prices_df.loc[:,'year-week'] = pd.to_datetime(prices_df.loc[:,'year-week'] + '-1', format='%Y-W%W-%w')
```

```python
prices_df = prices_df.drop(columns='yearweek').rename(columns={'year-week': 'yearweek'})
```

```python
# prices_df.head(3)
```

```python
# Crear columna id -> item + store_code -> para poder hacer merges y llevar el precio al ds principal
prices_df['id'] = prices_df['item'] + '_' + prices_df['store_code']
```

```python
# Obtener el ndice de la columna 'id'
id_index = prices_df.columns.get_loc('id')

# Insertar la columna 'id' al principio del dataframe despus del ndice
prices_df.insert(1, 'id', prices_df.pop('id'))
```

```python
# Nos quedamos solo con la columna id para poder mergear en el df de sales
prices_df = prices_df.drop(columns=['item','category','store_code'])
```

```python
# Ordenamos por ao y reseteamos el indice
prices_df = prices_df.sort_values('year', ascending=True).reset_index(drop=True)
```

```python
prices_df.head(3)
```

```python
# Guardamos prices_df para poder trabajar mas rapido
# prices_df.to_csv('data\prices_clean.csv', index=False)
```

```python
# Convertir la columna 'date' de merged_df y 'yearweek' de prices_df a tipo datetime
merged_df['date'] = pd.to_datetime(merged_df['date'])
prices_df['yearweek'] = pd.to_datetime(prices_df['yearweek'])

# Fusionar los dataframes basados en las fechas
final_df = pd.merge(merged_df, prices_df[['yearweek', 'week', 'sell_price']], how='left', left_on='date', right_on='yearweek')

# Eliminar la columna 'yearweek' ya que no es necesaria
final_df.drop(columns=['yearweek'], inplace=True)
```

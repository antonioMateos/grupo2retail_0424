```python
def ev_sales_time(df,l_title,l_x,l_y):
    # Evolucion total ventas por año
    plt.figure(figsize=(10, 6))
    df.plot(kind='bar', color='skyblue')
    plt.title(l_title)
    plt.xlabel(l_x)
    plt.ylabel(l_y)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Evolucion total ventas por año
ev_sales_time(total_ventas_por_año,'Evolucion total de ventas por año','Año','Total de ventas')
```

##### **2015 vs 2016**

```python
# Obtener df con total de ventas por mes para un año
def monthly_sales_per_year(target_year):
    # Convertir la columna 'date' a datetime -> Ya deberia estar
    merged_df['date'] = pd.to_datetime(merged_df['date'])

    # Añadir mes a partir de la columna 'date'
    merged_df['month'] = merged_df['date'].dt.month

    # Filtrar para obtener solo el año 2015
    temp_df = merged_df[merged_df['year'] == target_year]

    # Agrupar por mes y sumar las ventas diarias
    temp_df = temp_df.groupby('month')['daily_sales'].sum().reset_index()

    # Renombrar las columnas para mayor claridad
    temp_df.columns = ['month', 'total_sales']

    return temp_df
```

```python
monthly_sales_2015 = monthly_sales_per_year(2015)
monthly_sales_2016 = monthly_sales_per_year(2016)
```

```python
# Evolucion total de ventas en 2015
ev_sales_time(
    monthly_sales_2015,
    'Evolucion total de ventas en 2015',
    'Mes',
    'Total de ventas'
    )

ev_sales_time(
    monthly_sales_2016,
    'Evolucion total de ventas en 2016',
    'Mes',
    'Total de ventas'
    )
```

##### **Regiones Por año**

```python
def graph_barplot(data,xcol,ycol,l_x,l_y,hue,l_title,legend):
    # Configurar el estilo de Seaborn
    sns.set(style="whitegrid")

    # Crear el gráfico de barras agrupadas
    plt.figure(figsize=(10, 6))
    sns.barplot(data=data, x=xcol, y=ycol, hue=hue, palette='viridis')

    # Añadir título y etiquetas
    plt.title(l_title)
    plt.xlabel(l_x)
    plt.ylabel(l_y)
    plt.legend(title=legend)

    # Mostrar el gráfico
    plt.show()
```

```python
# Agrupar por año y región, y sumar las ventas
annual_sales_by_region = merged_df.groupby(['year', 'region'])['daily_sales'].sum().reset_index()

graph_barplot(
    annual_sales_by_region,
    'year',
    'daily_sales',
    'Año',
    'Total de ventas',
    'region',
    'Evolución del Total de Ventas Anual por Región',
    'Región'
    )
```

```python
# Filtrar los datos para la región de New York
ny_data = merged_df[merged_df['region'] == 'New York']

# Agrupar por año y tienda, y sumar las ventas
annual_sales_by_store = ny_data.groupby(['year', 'store'])['daily_sales'].sum().unstack()

# Crear el gráfico de barras apiladas
annual_sales_by_store.plot(kind='bar', figsize=(12, 6))

# Añadir título y etiquetas
plt.title('Evolución Anual del Total de Ventas para las Tiendas en la Región de New York')
plt.xlabel('Año')
plt.ylabel('Total de Ventas')
plt.legend(title='Tienda', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar el layout para evitar el corte de etiquetas
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Mostrar las tiendas identificadas como valores atípicos
# Calcular el volumen total de ventas mensual para cada tienda
monthly_sales_by_store = merged_df.groupby(['store', 'year', 'month'])['daily_sales'].sum().reset_index()

# Calcular la media y la desviación estándar del volumen total de ventas mensual para todas las tiendas
mean_sales = monthly_sales_by_store['daily_sales'].mean()
std_dev_sales = monthly_sales_by_store['daily_sales'].std()

# Definir los umbrales para identificar tiendas con volúmenes de ventas anormalmente grandes o pequeños
threshold_factor = 2  # Factor de multiplicación de la desviación estándar para definir los umbrales
upper_threshold = mean_sales + threshold_factor * std_dev_sales
lower_threshold = mean_sales - threshold_factor * std_dev_sales

# Identificar tiendas con volúmenes de ventas anormalmente grandes o pequeños
monthly_sales_by_store['outlier'] = monthly_sales_by_store['daily_sales'].apply(
    lambda x: 'up' if x > upper_threshold else 'down' if x < lower_threshold else 'normal')

print("Tiendas con volúmenes de ventas atípicamente grandes o pequeños:")
print(monthly_sales_by_store.sort_values('year',ascending=False))
```

#### **Categorias**

```python
# Global: Categorias que más / menos se han vendido historicamente
# Agrupar por año y categoría, sumar las ventas diarias y calcular las ventas totales anuales
category_sales_yearly = merged_df.groupby(['year', 'category'])['daily_sales'].sum().reset_index()
```

```python
# Crear un gráfico de líneas para mostrar la evolución anual de las ventas totales de las categorías
plt.figure(figsize=(12, 6))
for category in category_sales_yearly['category'].unique():
    plt.plot(category_sales_yearly[category_sales_yearly['category'] == category]['year'],
             category_sales_yearly[category_sales_yearly['category'] == category]['daily_sales'],
             label=category)

plt.title('Evolución anual de ventas totales por categoría (2011-2016)')
plt.xlabel('Año')
plt.ylabel('Ventas totales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Crear una tabla pivote para tener los años en las filas y las regiones en las columnas, con las ventas totales de cada categoría como valores
pivot_table = merged_df.pivot_table(index=['year', 'region'], columns='category', values='daily_sales', aggfunc='sum')
```

```python
# Crear un gráfico de barras agrupadas para mostrar la evolución de las ventas totales de categorías por año, región y categoría
plt.figure(figsize=(20, 12))
pivot_table.plot(kind='bar', stacked=True, cmap='viridis', edgecolor='w', linewidth=0.5)
plt.title('Evolución de ventas totales de categorías por año, región y categoría (2011-2016)')
plt.xlabel('Año - Región')
plt.ylabel('Ventas totales')
plt.legend(title='Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

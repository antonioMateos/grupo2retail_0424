```python
import pandas as pd
from sqlalchemy import create_engine
import sqlalchemy
import mysql

pd.set_option('display.float_format', lambda x: '%.2f' % x)
pd.set_option('display.max_columns',100)
pd.set_option('display.width',100)
pd.set_option('display.max_colwidth',100)
```

```python
# Cambiar por tus datos locales (estos son los de juan)
user = 'root'
password  = 'accountsandroles_99'
host= 'localhost'
bbdd = 'dsmarket'
cadena_conexion=f"mysql+mysqlconnector://{user}:{password}@{host}/{bbdd}"
```

```python
def get_df(table):
    try:
        motor = create_engine(cadena_conexion)
        consulta = f'SELECT * FROM {table}'
        df = pd.read_sql_query(consulta, motor)
        print(f"{table} cargada")
        return df
    except Exception as e:
        print(f"Ocurrió un error: {e}")
```

```python
df_tiendas = get_df('tiendas')
df_departamentos = get_df('departamentos')
df_articulos = get_df('articulos')
```

```python
# Contar el número de regiones únicas
num_regiones = df_tiendas['region'].nunique()
print(f"Número de regiones: {num_regiones}")
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')
# Contar el número de tiendas por región
tiendas_por_region = df_tiendas['region'].value_counts()
print("Número de tiendas por región:")
print(tiendas_por_region)
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')
# Mostrar los nombres de las regiones y las tiendas en cada región
print("Regiones y sus tiendas:")
for region in df_tiendas['region'].unique():
    tiendas = df_tiendas[df_tiendas['region'] == region]['store']
    print(f"{region}:")
    for tienda in tiendas:
        print(f" - {tienda}")
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')
# Contar el número total de tiendas
num_tiendas_total = df_tiendas['store_code'].nunique()
print(f"Número total de tiendas: {num_tiendas_total}")
```

```python
# Contar el número de categorías únicas
num_categorias = df_departamentos['category'].nunique()
print(f"Número de categorías: {num_categorias}")
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')

# Contar el número de departamentos por categoría
departamentos_por_categoria = df_departamentos['category'].value_counts()
print("Número de departamentos por categoría:")
print(departamentos_por_categoria)
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')

# Mostrar los nombres de las categorías y los departamentos en cada categoría
print("Categorías y sus departamentos:")
for categoria in df_departamentos['category'].unique():
    departamentos = df_departamentos[df_departamentos['category'] == categoria]['department']
    print(f"{categoria}:")
    for departamento in departamentos:
        print(f" - {departamento}")
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')

# Contar el número total de departamentos
num_departamentos_total = df_departamentos['department'].nunique()
print(f"Número total de departamentos: {num_departamentos_total}")
```

```python
# Contar el número total de artículos distintos
num_articulos_total = df_articulos['item'].nunique()
print(f"Número total de artículos distintos: {num_articulos_total}")
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')

# Contar el número de artículos distintos por categoría
articulos_por_categoria = df_articulos.groupby('category')['item'].nunique()
print("Número de artículos distintos por categoría:")
print(articulos_por_categoria)
print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - ')

# Contar el número de artículos distintos por departamento
articulos_por_departamento = df_articulos.groupby('department')['item'].nunique()
print("Número de artículos distintos por departamento:")
print(articulos_por_departamento)
```

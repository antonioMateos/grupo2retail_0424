## TRATAMIENTO DE DATOS

__Hacemos `melt` con las columnas de fechas para tener un registro por fecha, item y tienda__

```python
df = df.melt(id_vars=['id','item','category','department','store','store_code','region'], var_name='day', value_name='units')

print(f'Tamaño df: {df.shape}')

df.head(3)
```

__Obtenemos fechas desde df_calendar__

```python
# Hacer el merge utilizando la columna 'day' de df y 'd' de df_cal
df_with_dates = df.merge(df_cal[['d', 'date']], left_on='day', right_on='d', how='left')

df_with_dates.head(3)
```

```python
print(f'Tamaño df: {df.shape}')
```

__Transformamos en `Datetime` la columna `date`__

```python
df_with_dates['date'] = pd.to_datetime(df_with_dates['date'], format = '%Y-%m-%d')
```

```python
# Copiamos para continuar
df = df_with_dates.copy()
```

__CREAMOS EL DATAFRAME `tiendas`__

```python
df_tiendas = df[['store_code', 'region', 'store']].drop_duplicates()
df_tiendas.reset_index(drop=True)
```

__Borramos las columnas que nos sobran para ir aligerando el df__

```python
df.drop(columns=['id'], inplace=True)
df.drop(columns=['region'], inplace=True)
df.drop(columns=['store'], inplace=True)
```

__CREAMOS EL DATAFRAME `departamentos`__

```python
df_departamentos = df[['category', 'department']].drop_duplicates()
df_departamentos.reset_index(drop=True)
```

__CREAMOS EL DATAFRAME `articulos`__

```python
df_articulos = df[['item', 'category', 'department']].drop_duplicates()
df_articulos.reset_index(drop=True)
```

__Borramos las columnas que nos sobran para ir aligerando el df__

```python
df.drop(columns=['category'], inplace=True)
df.drop(columns=['department'], inplace=True)
```

__BORRAMOS TODAS LAS FILAS CUYAS VENTAS SEAN 0__

```python
print(f'Tamaño df antes borrar 0 unidades: {df.shape}')
df = df[df['units'] != 0]
print(f'Tamaño df despues borrar 0 unidades: {df.shape}')
```

__CREAMOS LAS COLUMNAS `year` Y `week` PARA CREAR TABLAS RESUMEN MAS LIBIANAS__

```python
df['year'] = df['date'].dt.year
#df['week'] = df['date'].dt.weekofyear
#df['week'] = df['date'].dt.dayofyear // 7 + 1
df['week'] = df['date'].dt.isocalendar().week

df.head(3)
```

__CREAMOS EL DATAFRAME RESUMEN POR weekS__

```python
df_week = df.groupby(['item', 'store_code', 'year', 'week'])['units'].sum().reset_index()

print(f'DF week:\n{df_week.shape}')
print('- - - - - -')
df_week = df_week.sort_values('year', ascending=True)
df_week.head(3)
```

__LEEMOS LA TABLA DE PRECIOS__

```python
df_precios = pd.read_csv(raw_data_path+'item_prices.csv')
```

```python
df_precios.head(3)
```

```python
df_precios.isna().sum()
```

```python
df_precios['yearweek'] = df_precios['yearweek'].fillna(200001)
df_precios.head(3)
```

```python
df_precios['yearweek'] = df_precios['yearweek'].astype(int).astype(str)

# Crear columnas 'año' y 'week'
df_precios['year'] = df_precios['yearweek'].str[:4].astype(int)
df_precios['week'] = df_precios['yearweek'].str[4:].astype(int)

# Eliminar la columna original si no se necesita
df_precios = df_precios.drop(columns=['yearweek'])
df_precios = df_precios.drop(columns=['category'])

print(f'DF_Precios cols: {df_precios.columns}')

df_precios.head(3)
```

```python
df_weekly_sales = df_week.merge(df_precios, on=['item', 'store_code', 'year', 'week'], how='left')
```

```python
print(f'DF Weekly Sales tamaño: {df_weekly_sales.shape}')

print(f'DF Weekly Sales sell_price = Nan: {df_weekly_sales.sell_price.isnull().sum()}')
```

```python
df_sell_price_nan = df_weekly_sales.loc[(df_weekly_sales['sell_price'].isna() )]
df_sell_price_nan.shape
```

__Creamos df de ventas por dia__

```python
df_day_sales = df.join(df_weekly_sales.set_index(['item', 'store_code', 'year', 'week'])[['sell_price']], on=['item', 'store_code', 'year', 'week'])
df_day_sales
```

```python
df_day_sales['raw_earn'] = df_day_sales['units']*df_day_sales['sell_price']
```

```python
df_weekly_sales['raw_earn'] = df_weekly_sales['units']*df_weekly_sales['sell_price']
```

```python
df_day_sales['month'] = df_day_sales['date'].dt.month
```

```python
df_month_sales = df_day_sales.groupby(['item', 'store_code', 'year', 'month']).agg({
    'units': 'sum',
    'raw_earn': 'sum'
}).reset_index()
```

```python
print(f'DF month_sales tamaño: {df_month_sales.shape}')
df_month_sales.head(3)
```

```python
df_year_sales = df_month_sales.groupby(['item', 'store_code', 'year']).agg({
    'units': 'sum',
    'raw_earn': 'sum'
}).reset_index()
```

```python
print(f'DF year_sales tamaño: {df_year_sales.shape}')
df_year_sales.head(3)
```

## Conectamos a `ds_market`

```python
import os
import pymongo
from dotenv import load_dotenv

# Cargar las variables de entorno desde el archivo .env
load_dotenv()

# Obtener la URI de MongoDB desde las variables de entorno
URI = os.getenv('LOCAL.URI')

# print(f'URI: {URI}')
# Verificar si la URI se ha cargado correctamente
if not URI:
    raise ValueError("No se ha encontrado la URI de MongoDB en las variables de entorno")

# Conectar a la base de datos de MongoDB
client = pymongo.MongoClient(URI)

# Crear / Conectar a DB
db = client['ds_market']
# db = client.get_database() # Si ya estuviera definida

# Verificar la conexión imprimiendo la lista de colecciones
if db is not None:
    print(':::: Conectado ::::')
```

```python
# Insertar cada coleccion en la db
def insert_dataframe_to_mongo(collection_name, df):
    """
    Inserta un DataFrame de pandas en una colección de MongoDB.

    Args:
    collection_name (str): El nombre de la colección en la que se insertarán los datos.
    df (pd.DataFrame): El DataFrame de pandas que se va a insertar en la colección.
    """
    # Crear o seleccionar la colección
    collection = db[collection_name]

    # Convertir el DataFrame a una lista de diccionarios
    data_dict = df.to_dict('records')

    # Insertar los datos en la colección
    collection.insert_many(data_dict)

    print(f"Datos insertados correctamente en la colección '{collection_name}'")
```

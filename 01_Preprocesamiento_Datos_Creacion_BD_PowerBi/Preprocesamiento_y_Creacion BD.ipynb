__CAMBIO LOS NOMBRES DE LAS COLUMNAS POR VALORES FECHAS__   
Según el calendario el día 1 es `2011-01-29`

```python
f = pd.Timestamp("2011-01-28")
for i in range(1913):
    i += 1
    f = f + pd.DateOffset(days=1)
    col = "d_"+ str(i)
    new_col = f.strftime('%Y/%m/%d')
    df.rename(columns={col: new_col}, inplace=True)
```

__Hacemos `melt` con las columnas de fechas para tener un registro por fecha, item y tienda__

```python
df = df.melt(id_vars=['id','item','category','department','store','store_code','region'], var_name='fecha', value_name='unidades')
df.shape
```

__Transformamos en `Datetime` la columna `fecha`__

```python
df['fecha'] = pd.to_datetime(df['fecha'])
```

### __Crea DF `tiendas`__

```python
df_tiendas = df[['store_code', 'region', 'store']].drop_duplicates()
df_tiendas.reset_index(drop=True)
```

```python
df.drop(columns=['id'], inplace=True)
df.drop(columns=['region'], inplace=True)
df.drop(columns=['store'], inplace=True)
```

### __Crea DF `categorias`__

```python
df_categorias = df[['category']].drop_duplicates()
df_categorias.reset_index(drop=True)
```

### __Crea DF Departamentos `departamentos`__

```python
df_departamentos = df[['category', 'department']].drop_duplicates()
df_departamentos.reset_index(drop=True)
```

### __Crea DF `articulos`__

```python
df_articulos = df[['item', 'category', 'department']].drop_duplicates()
df_articulos.reset_index(drop=True)
```

### Borrado columnas

__Borramos las columnas que nos sobran para ir aligerando el df__

```python
df.drop(columns=['category'], inplace=True)
df.drop(columns=['department'], inplace=True)
```

__BORRAMOS TODAS LAS FILAS CUYAS VENTAS SEAN 0__

```python
df = df[df['unidades'] != 0]
df
```

### __Cols `ejercicio` Y `semana` PARA CREAR TABLAS RESUMEN MAS LIBIANAS__

```python
df['ejercicio'] = df['fecha'].dt.year
#df['semana'] = df['fecha'].dt.weekofyear
#df['semana'] = df['fecha'].dt.dayofyear // 7 + 1
df['semana'] = df['fecha'].dt.isocalendar().week
df
```

### __DF RESUMEN POR SEMANAS__

```python
df_semana = df.groupby(['item', 'store_code', 'ejercicio', 'semana'])['unidades'].sum().reset_index()
```

### __TABLA DE PRECIOS__

```python
df_precios = pd.read_csv(raw_data_path+"item_prices.csv")
```

```python
print (df_precios.yearweek.value_counts(dropna=False).sort_values())
```

```python
df_precios
```

```python
df_precios['yearweek'] = df_precios['yearweek'].fillna(200001)
df_precios
```

```python
df_precios['yearweek'] = df_precios['yearweek'].astype(int).astype(str)

# Crear columnas 'año' y 'semana'
df_precios['ejercicio'] = df_precios['yearweek'].str[:4].astype(int)
df_precios['semana'] = df_precios['yearweek'].str[4:].astype(int)

# Eliminar la columna original si no se necesita
df_precios = df_precios.drop(columns=['yearweek'])
df_precios = df_precios.drop(columns=['category'])
df_precios
```

```python
df_precios.columns
```

```python
df_precios.head(10)
```

```python
df_semana.head(10)
```

```python
df_vensem = df_semana.merge(df_precios, on=['item', 'store_code', 'ejercicio', 'semana'], how='left')
```

```python
len(df_vensem)
```

```python
df_vensem.sell_price.isnull().sum()
```

```python
df_nulos = df_vensem.loc[(df_vensem['sell_price'].isna() )]
df_nulos
```

```python
df_vensem = df_vensem.sort_values(by=['item', 'store_code', 'ejercicio', 'semana'])
df_vensem['sell_price'] = df_vensem.groupby(['item', 'store_code', 'ejercicio'])['sell_price'].bfill()
df_vensem['sell_price'] = df_vensem.groupby(['item', 'store_code', 'ejercicio'])['sell_price'].ffill()
df_vensem['sell_price'] = df_vensem.groupby(['item', 'store_code'])['sell_price'].ffill()
df_vensem['sell_price'] = df_vensem.groupby(['item', 'store_code'])['sell_price'].bfill()
```

```python
df_nulos = df_vensem.loc[(df_vensem['sell_price'].isna() )]
df_nulos
```

### __DF Ventas Diarias__

```python
df_vendia = df.join(df_vensem.set_index(['item', 'store_code', 'ejercicio', 'semana'])[['sell_price']], on=['item', 'store_code', 'ejercicio', 'semana'])
df_vendia
```

```python
df_vendia['cifra'] = df_vendia['unidades']*df_vendia['sell_price']
```

```python
df_vensem['cifra'] = df_vensem['unidades']*df_vensem['sell_price']
```

```python
df_vendia['mes'] = df_vendia['fecha'].dt.month
```

### __DF Ventas Mensuales__

```python
df_venmes = df_vendia.groupby(['item', 'store_code', 'ejercicio', 'mes']).agg({
    'unidades': 'sum',
    'cifra': 'sum'
}).reset_index()
```

```python
df_venmes
```

### __DF Ventas Anuales__

```python
df_venano = df_venmes.groupby(['item', 'store_code', 'ejercicio']).agg({
    'unidades': 'sum',
    'cifra': 'sum'
}).reset_index()
```

```python
df_venano
```
